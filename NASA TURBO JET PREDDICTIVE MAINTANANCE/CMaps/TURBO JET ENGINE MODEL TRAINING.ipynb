### COMPUTING REMAINING USEFULL LIFE OF THE ENGINE


```python
max_cycles_df=raw_data.groupby(['ID'],sort=False)['CYCLE'].max().reset_index().rename(columns={'CYCLE':'MAXCYCLE_ID'})
max_cycles_df.head()
```

```python
max_cycles_df
```

```python
# merging the max cycle and raw data to calculate RUL
FD001_df=pd.merge(raw_data,max_cycles_df,how='inner',on='ID')
```

```python
FD001_df
```

```python
# to calculate RUL we have subtract cycles - maxcycle_id
FD001_df['RUL']=FD001_df['MAXCYCLE_ID']-FD001_df['CYCLE']
FD001_df
```

## Correlation Matrix with Heatmap
Correlation states how the features are related to each other or the target variable.

Correlation can be positive (increase in one value of feature increases the value of the target variable) or negative (increase in one value of feature decreases the value of the target variable)

Heatmap makes it easy to identify which features are most related to the target variable, we will plot heatmap of correlated features using the seaborn library.

```python
scaler =StandardScaler()

X_scaled = scaler.fit_transform(X_DATA)
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
variables = X_scaled

# we create a new data frame which will include all the VIFs
# note that each variable has its own variance inflation factor as this measure is variable specific (not model specific)
# we do not include categorical values for mulitcollinearity as they do not provide much information as numerical ones do
vif = pd.DataFrame()

# here we make use of the variance_inflation_factor, which will basically output the respective VIFs 
vif["VIF"] = [variance_inflation_factor(variables, i) for i in range(variables.shape[1])]
# Finally, I like to include names so it is easier to explore the result
vif["Features"] = X_DATA.columns
```

```python
x_train,x_test,y_train,y_test = train_test_split(X_scaled,Y_DATA,test_size = 0.25,random_state=355)
```

```python
x_test
```

```python
regression = LinearRegression()

regression.fit(x_train,y_train)
```

```python
regression.score(x_train,y_train)
```

```python
def adj_r2(x,y):
    r2 = regression.score(x,y)
    n = x.shape[0]
    p = x.shape[1]
    adjusted_r2 = 1-(1-r2)*(n-1)/(n-p-1)
    return adjusted_r2
```
