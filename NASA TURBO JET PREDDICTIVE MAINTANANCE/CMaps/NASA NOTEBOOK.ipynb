### Classification Problem Formulation
    The challange of this project was to predict the Remaining Useful Life of the engine by using the given sensor's data and operational conditions.

    But in this project, we ll try to simplify that by converting it to a Classification Problem. Where, the classes/labels will of 3 types, i.e. Good Condition, Moderate Condition and Warning Condition.

    labels corresponding to each conditions
        Good Condition - 2
        Moderate Condition - 1
        Warning Condition - 0
    Defining the labels
    Here we will define the engine's condition with RUL	, which is the difference between Current Cycle and the End cycle. If RUL=2, that means the component has just started its degradation and RUL=0 means, it is completely degraded.

    if RUL <= 80 - Warning Condition
    if 80 < RUL <= 180 - Moderate Condition
    if 180 < RUL - Good Condition

```python
# Creating the lable coloums with the RUL conditions
labels=[]
for i in range (0,len(FD001_df)):
    if np.array(FD001_df["RUL"])[i] <= 80: # less than 80 RUL Warning condition
        labels.append(0)
    elif np.array(FD001_df["RUL"])[i] <= 180: # less than are equal to 180 RUL Warning condition
        labels.append(1)   
    else :
        labels.append(2)  # greater than  180 RUL Warning condition
        
FD001_df["labels"]=labels
FD001_df  = FD001_df.drop(columns=['RUL'])   
FD001_df
```
